##  Step 2: Importing and Loading the Model

Now that Whisper is installed, we need to do two things:

---

##  Step 3: Transcribing Audio 

### What is transcription?

**Transcription** means converting spoken words into written text. When you listen to a podcast and someone writes down everything that was said — that person just did transcription. Our AI model does this automatically.

### What happens behind the scenes?

When you call `model.transcribe("audio.wav")`, here's what happens :

1. **Load audio:** Whisper reads the audio file using FFmpeg
2. **Convert to spectrogram:** The audio waveform is converted into a visual representation called a **log-Mel spectrogram** 
3. **Encoder processes it:** The encoder part of Whisper analyzes the spectrogram and creates an internal understanding of the audio
4. **Decoder generates text:** The decoder takes that understanding and predicts text, one word at a time
5. **Return results:** The transcribed text is returned to you!

### What does `result["text"]` mean?

The `transcribe()` function returns a **dictionary** . The dictionary contains several pieces of information:

| Key | What it contains |
|---|---|
| `result["text"]` | The full transcribed text |
| `result["segments"]` | Text broken into time-stamped segments |
| `result["language"]` | The detected language of the audio |


```python
# ============================================
# TRANSCRIPTION: Converting speech to text!
# ============================================

result = model.transcribe("audio.wav")

# Print the transcribed text
print("="*60)
print(" TRANSCRIPTION RESULT (Base Model)")
print("="*60)
print()
print("Transcribed text:")
print(result["text"])
print()
print(f"Detected language: {result['language']}")
print("="*60)
```

###  Understanding the Output Above

The above output is what the AI **thinks** was said in the audio file!

**Is it perfect?** Maybe, maybe not. STT models are very good but not 100% accurate. Here are things that can affect accuracy:

- **Background noise** — Music, traffic, or other sounds confuse the model
- **Accent/pronunciation** — Strong accents can cause errors
- **Recording quality** — Low-quality microphones reduce accuracy
- **Fast speech** — Speaking very quickly makes it harder
- **Rare words** — Unusual names or technical terms may be wrong

Now let's try the same audio with different model sizes to see how they compare!

## Experiment 1: The Tiny Model (Speed Test)

### Why are we doing this?

We want to see what happens when we use a **smaller, faster** model. The tiny model has:
- Only **39 million** parameters (compared to 74M for base)
- **Much faster** processing speed
- **Less memory** required

### What should we expect?

Since the tiny model has fewer "brain cells," it should be:
- ✅ **Faster** at processing
- ❌ **Less accurate** in its transcription


```python
# ============================================
# EXPERIMENT 1: Tiny Model
# ============================================

import time  # To measure how long it takes

# Load the tiny model (39 million parameters)
# This is the smallest Whisper model available
print("Loading tiny model...")
model_tiny = whisper.load_model("tiny")
print(" Tiny model loaded!\n")

# Measure how long transcription takes
start_time = time.time()  # Record the start time
result_tiny = model_tiny.transcribe("audio.wav")  
end_time = time.time()  # Record the end time

# Calculate elapsed time
elapsed = round(end_time - start_time, 2)

# Display results
print("="*60)
print(" TRANSCRIPTION RESULT (Tiny Model)")
print("="*60)
print()
print("Transcribed text:")
print(result_tiny["text"])
print()
print(f" Time taken: {elapsed} seconds")
print(f"Detected language: {result_tiny['language']}")
print("="*60)
```

###  Analysis: Tiny Model

Compare the tiny model's output with the base model's output above.

**Things to look for:**
- Did it get all the words right?
- Were any words skipped or added?
- Was it faster than the base model?

The tiny model is great for situations where you need **speed** over perfection — like real-time captioning or quick drafts.

##  Experiment 2: The Small Model (Accuracy Test)

### Why are we doing this?

Now let's go the other direction — use a **bigger, more powerful** model. The small model has:
- **244 million** parameters (over 6x more than tiny!)
- **More accuracy** but requires more time and memory

### What should we expect?

Since the small model has more "brain cells," it should be:
- ✅ **More accurate** — catches words that tiny missed
- ❌ **Slower** — takes longer to process
- ❌ **Uses more memory** — needs ~2 GB of RAM



```python
# ============================================
# EXPERIMENT 2: Small Model
# ============================================

# Load the small model (244 million parameters)

print("Loading small model (this may take a minute)...")
model_small = whisper.load_model("small")
print(" Small model loaded!\n")

# Measure transcription time
start_time = time.time()
result_small = model_small.transcribe("audio.wav")
end_time = time.time()

elapsed = round(end_time - start_time, 2)

# Display results
print("="*60)
print(" TRANSCRIPTION RESULT (Small Model)")
print("="*60)
print()
print("Transcribed text:")
print(result_small["text"])
print()
print(f" Time taken: {elapsed} seconds")
print(f"Detected language: {result_small['language']}")
print("="*60)
```

### Analysis: Small Model

Now compare **all three** results:

1. **Tiny** → Fastest, but might miss or change some words
2. **Base** → Good balance of speed and accuracy
3. **Small** → Most accurate, but slowest

This is a fundamental concept in AI called the **speed-accuracy trade-off**: you can have fast OR accurate, but getting both at the same time is expensive.

##  Step 5: Exploring Segments (Timestamps)

Whisper doesn't just give you text — it also tells you **when** each word was spoken! This is called **segmentation**.

### What are segments?

A segment is a chunk of text with a start time and end time. For example:
- `[0.00s → 2.50s]` "Hello, welcome to the presentation"
- `[2.50s → 5.30s]` "Today we will talk about speech recognition"

This is how YouTube generates those **timed subtitles** you see on videos!


```python
# ============================================
# EXPLORING SEGMENTS (TIMESTAMPS)
# ============================================

# The 'segments' key contains time-stamped text chunks
# Each segment has: start time, end time, and text

print("="*60)
print(" TIMESTAMPED SEGMENTS (Base Model)")
print("="*60)
print()

for i, segment in enumerate(result["segments"]):
    start = round(segment["start"], 2)  # Start time in seconds
    end = round(segment["end"], 2)      # End time in seconds
    text = segment["text"]               # The text for this segment
    
    print(f"  Segment {i+1}: [{start}s → {end}s]")
    print(f"  Text: {text}")
    print()
```

###  What did we just see?

Each segment shows:
- **Start time** → When this piece of speech begins (in seconds)
- **End time** → When it ends
- **Text** → What was said during that time window

This is incredibly useful for:
- Creating **subtitles** for videos
- **Searching** through long audio 
- **Summarizing** specific parts of a recording

##  Challenges We Faced

Here are the problems and limitations we encountered during this project:

1. **Background noise ruins accuracy** — When the audio has music, traffic, or other voices, the model makes more mistakes. It sometimes even 'hallucinates' (generates words that were never spoken).

2. **Accents can be tricky** — Non-standard accents occasionally confuse the model, especially the smaller versions.

3. **Big models need big hardware** — The medium and large models require expensive GPUs with lots of memory. On free-tier services (like Google Colab), we can only reliably use tiny, base, and small.

4. **FFmpeg dependency** — Whisper requires FFmpeg to be installed separately, which can be confusing for beginners.
